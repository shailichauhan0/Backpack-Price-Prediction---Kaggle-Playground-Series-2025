```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.preprocessing import LabelEncoder
```

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
train_df.head()
```

```python
test_df.head()
```

```python
train_df.describe()
```

```python
encoder = LabelEncoder()
```

```python
train_df['Material'] = encoder.fit_transform(train_df['Material'])
train_df['Laptop Compartment'] = encoder.fit_transform(train_df['Laptop Compartment'])
train_df['Size'] = encoder.fit_transform(train_df['Size'])
train_df['Style'] = encoder.fit_transform(train_df['Style'])
train_df['Waterproof'] = encoder.fit_transform(train_df['Waterproof'])
train_df['Color'] = encoder.fit_transform(train_df['Color'])
train_df['Brand'] = encoder.fit_transform(train_df['Brand'])
```

```python
train_df
```

```python
train_df['Weight Capacity (kg)'] = train_df['Weight Capacity (kg)'].round(2)
train_df['Price'] = train_df['Price'].round(3)
train_df
```

```python
train = pd.read_csv('train.csv')
train.head()
```

```python
for col in train.columns[:-2]:
    train[col] = encoder.fit_transform(train[col])

train.head()
```

```python
for col in train.columns[-2:]:
  train[col] = train[col].round(3)
```

```python
train['Brand'].unique()
```

```python
train.head(-1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
def transform(dataframe):
  for col in dataframe.columns[:-1]:
      dataframe[col] = encoder.fit_transform(dataframe[col])

  for col in dataframe.columns[-2:]:
    dataframe[col] = dataframe[col].round(3)
```

```python
transform(test_df)
test_df.head(-1)
```

```python
train['Weight Capacity (kg)'] = train['Weight Capacity (kg)'].fillna(train['Weight Capacity (kg)'].mean())
```

```python
x_train = train[train.columns[:-1]]
y_train = train[train.columns[-1:]]
x_test = test_df
```

```python
x_train.isnull().sum()
```

```python
x_test.isnull().sum()
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
test_df.isnull().sum()
```

```python
test_df['Weight Capacity (kg)'] = test_df['Weight Capacity (kg)'].fillna(test_df['Weight Capacity (kg)'].mean())
```

```python
test_df.isnull().sum()
```

```python
predictions = model.predict(x_test)
```

```python
predictions = pd.DataFrame(predictions, columns=['Price'])
```

```python
predictions['id'] = predictions.index
```

```python
predictions = predictions[['id', 'Price']]
```
